# Data Visulization

```python
sns.barplot(x='year',y='AveragePrice',data=df,hue='type')
```

```python
fig,ax=plt.subplots()
sns.boxplot(x='type',y='AveragePrice',data=df)
```

```python
df.plot(kind = 'box',subplots = True,layout = (4,3),sharex = False, sharey=False,figsize=(10,10))
plt.show()
```

```python
def plot_count(x,fig):
    
    plt.subplot(4,2,fig)
    plt.title(x+' vs small Hass')
    sns.scatterplot(x=x,y='small Hass',data=df)
    
```

```python
sns.scatterplot(x='small Hass',y=df['large Hass'],data=df)
```

```python

plt.figure(figsize=(20,20))
plot_count('large Hass',1)
plot_count('Small Bags',3)
plot_count('Large Bags',5)
plot_count('AveragePrice',7)
```
